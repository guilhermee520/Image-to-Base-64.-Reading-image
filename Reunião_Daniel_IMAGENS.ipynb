```python
# NUMPY para transforma imagem em Array e mostrar a Matriz
imagem = np.array(im)
```

```python
# SETANDO O DICIONÁRIO COM __CHAVE__VALOR__
width = len(imagem)
height = len(imagem[0])   

lista = {}

for i in range(width):
    for a in range(height):
        #print (f'altura: {a} largura: {i} {arrimg[a][i]}' )
        #json_data = f'{a} {i} {arrimg[i][a]}'
        json_data = {f'{a}_{i}' :f'{imagem[i][a]}'}
        lista.update(json_data)
```

```python
# CRIANDO JSON .TXT -> COM __CHAVE__VALOR__ 
with open ("teste_chave_valor.txt","w" ) as escritor:
    json.dump(lista,escritor)
escritor.close()
```

# aprendendo string -> para -> base64

```python
# TRANFORMANDO STRING EM BASE64

#  STRING -> PARA -> BYTES = ENCODE
#  BYTES -> PARA -> STRING = DECODE
import base64
string_b = "Gustavo de Araujo".encode('utf-8')

base64_bytes = base64.b64encode(string_b)

print(base64_bytes)
```

```python
#  VOLTANDO DE BASE64 -> PARA -> STRING
base64_decoded_bytes = base64.decodebytes(base64_bytes)
print (base64_decoded_bytes.decode())
```
